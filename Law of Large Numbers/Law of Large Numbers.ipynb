### Introduction
In probability theory, this law states that the average of the results obtained from a large number of trials should be closer to the expected value and as more trials are performed.
For example, as a sample size of a coin flip grows, its average of either heads/tails gets closer to the true probability of 50%. Therefore, the probability of 2 heads in a row should be 25% given a large enough trails. Let's prove this theory out.
### Analysis

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np

# create function with 2 coins that contains heads and tails
def coin_flip_experiment():
  
  # defining our two coins as lists
  coin1 = ['Heads', 'Tails']
  coin2 = ['Heads', 'Tails']
 
  # "flipping" both coins randomly, np.random.choice picks random element (heads or tails) from coin1 and coin2
  coin1_result = np.random.choice(coin1)
  coin2_result = np.random.choice(coin2)
 
  # checking if both flips are heads, return 1 if it is
  if coin1_result == 'Heads' and coin2_result == 'Heads':
    return 1
  else:
    return 0
```

Now that we have our coinflip function set up, let's create the experiment. We want a large sample size, so we want to find the percentage of flips that both coins landed on heads after 1000 flips. To find the current probability, we need to divide the number of times we get 2 heads and the total amount of flips. To visualize this, we will create a probability list that holds the most recent 2 head percentage after every flip.   

```python

# how many times we run the experiment
num_trials = 1000
# keeps a list of the probability of two heads at each flip, should get closer to true probability of 0.25 as more flips go by
probability = []
# keeps track of number of trails ran
flips = []
# keep track of the number of times heads pops up twice
two_heads_counter = 0
 
# perform the experiment X times
for flip in range(num_trials):
    
  # if both coins are heads, add 1 to the counter
  two_heads_counter += coin_flip_experiment()

  # counter / number of times flipped = probability of two heads so far, append each updated probability to list
  probability.append(two_heads_counter/(flip+1))
  
  # keep a list for number of flips, goes until num_trail (1000)
  flips.append(flip+1)

# we can see our first flip was tail then head and then it starts getting closer to .25 
print(probability[0:11])
 
# plot all flips and probability of two heads
plt.plot(flips, probability, label='Experimental Probability')
plt.xlabel('Number of Flips')
plt.ylabel('Proportion of Two Heads')

# lets add an orange line at .25 to signify the expected probability
plt.hlines(0.25, 0, num_trials, colors='orange', label='True Probability')
plt.legend()

plt.show()
```

### Conclusion
As we can see, the probability after the first few flips are extremely unpredictable but as the Law of Large Numbers states: trials should be closer to the expected value and as more trials are performed and that is exactly what we see here after 1000 flips.
